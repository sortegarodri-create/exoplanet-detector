```python
#importar librería
import lightkurve as lk
```

```python
#buscar la estrella
resultado_busqueda = lk.search_lightcurve('KIC 3733346', author = 'Kepler')
resultado_busqueda
```

```python
#descargar la curva de luz
lc = resultado_busqueda.download()
lc
```

```python
#mostrar la curva
lc.plot()
```
